```python
from google.colab import drive
import sys
import os

print("Mounting Google Drive...")
drive.mount('/content/drive')

# Define the path to your project folder
PROJECT_PATH = '/content/drive/MyDrive/Colab Notebooks/CV_FALL_2025/Final Project'

# Add the project and modules path to Python's system path
sys.path.append(PROJECT_PATH)
sys.path.append(os.path.join(PROJECT_PATH, 'modules'))

# Change the current working directory to your project folder
%cd {PROJECT_PATH}

# Set up Hugging Face cache to use your Google Drive for persistent model storage
MODELS_PATH = os.path.join(PROJECT_PATH, 'models')
os.makedirs(MODELS_PATH, exist_ok=True) # Ensure the directory exists
os.environ['HF_HOME'] = MODELS_PATH
os.environ['TRANSFORMERS_CACHE'] = MODELS_PATH
os.environ['DIFFUSERS_CACHE'] = MODELS_PATH

print("-" * 50)
print(f"Current working directory: {os.getcwd()}")
print(f"Hugging Face model cache set to: {MODELS_PATH}")
print("Environment setup complete.")
print("-" * 50)
```
